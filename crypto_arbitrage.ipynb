#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange).
```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

#Calculating the returns by dividing the difference in price on positive arbitrages by the buying price early on.
spread_return_early = arbitrage_spread_early[arbitrage_spread_early>0]/bitstamp_sliced.loc["2018-1-02":"2018-1-02"]

# Review the spread return DataFrame
# YOUR CODE HERE

#Describing the resulting dataframe to get more data.
spread_return_early.describe()
```

```python
#Calculating the returns by dividing the difference in price on positive arbitrages by the buying price in the middle.
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle>0]/bitstamp_sliced.loc["2018-2-15":"2018-2-15"]

#Describing the resulting dataframe to get more data.
spread_return_middle.describe()
```

```python
#Calculating the returns by dividing the difference in price on positive arbitrages by the buying price later on.
spread_return_late = arbitrage_spread_late[arbitrage_spread_late>0]/bitstamp_sliced.loc["2018-3-30":"2018-3-30"]

#Describing the resulting dataframe to get more data.
spread_return_late.describe()
```
